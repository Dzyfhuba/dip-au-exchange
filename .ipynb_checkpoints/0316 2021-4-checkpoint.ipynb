```python
from PIL import Image # for Gaussian blur
```

```python
from PIL import ImageFilter
```

```python
img = Image.open("E:/Computer Vision/newastronaut.png")  
```

```python
blur_img = img.filter(ImageFilter.GaussianBlur(5))
```

```python
blur_img.show()
```

```python
from skimage import io, filters
```

```python
img = io.imread("E:/Computer Vision/newastronaut.png")
```

```python
out = filters.gaussian(img, sigma=5)
```

```python
io.imshow(out)
```

```python
from skimage import io # for edge detection
```

```python
from skimage import color
```

```python
from skimage import filters
```

```python
img = io.imread("E:/Computer Vision/astronaut.png")
```

```python
img = color.rgb2gray(img)
```

```python
edge = filters.sobel(img)
```

```python
io.imshow(edge)
```
