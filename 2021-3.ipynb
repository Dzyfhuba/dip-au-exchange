```python
from PIL import Image
```

```python
img = Image.open("newastronaut.png")
```

```python
img.getpixel((100,100))
```

```python
width, height = img.size 
```

```python
for x in range(width): 
    img.putpixel( (width-x-1, x), (0, 0, 255) ) 
    img.putpixel( (x, x), (0, 255, 0) )
    img.putpixel( (int(width/2), int(x)), (255, 0, 0))
    img.putpixel( (int(x), int(height/2)), (255, 255, 255))
```

```python
# img.show()
display(img)
```
